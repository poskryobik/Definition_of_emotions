```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from PIL import Image
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy
```

```python
#Создание класса
class emotion_on_camera():
    '''
        Класс который изспользует обученную модель checkpoint_best.h5
        и готовое решение библиотеки OpenCV.
    '''
    def __init__(self):
        #Загружаем модель
        self.model = load_model('checkpoint_best.h5')
        #Список классов
        self.classes = ['anger', 'contempt', 'disgust', 'fear', 'happy', 
                        'neutral', 'sad', 'surprise', 'uncertain']
        #Объект камера
        self.cam = cv2.VideoCapture(0)
        # иницилизируем детектор:
        self.face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
        
    #Проверка работоспособности камеры
    def check_work_cam(self):
        if not self.cam.isOpened():
            print("Не удалось открыть камеру")
        else:
            print("Камера запущена")
    #Получение изображения
    def get_frame(self):
        ret, self.frame = self.cam.read()
        if ret:
            return self.frame
    
    
    
    #Предсказание модели на части изображения
    def predict(self, img):
        answer = self.model.predict(img)
        res_class = list(answer[0]).index(answer[0].max()) 
        return self.classes[res_class]
    
    #Классы эмоций на изображение
    def add_bounding_box(self):
        frame = self.get_frame()
        #Приведение к ЧБ
        grayscale_image = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        #Определение лиц
        faces = self.face_detector.detectMultiScale(grayscale_image, 1.3, 5)
        #Приведение к RGB
        rgb_image_with_boundingbox = deepcopy(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        for i, bounding_box in enumerate(faces):
            x, y, w, h = bounding_box
            #Находим лицо
            face_boundingbox_bgr = frame[y:y + h, x:x + w]
            face_boundingbox_rgb = cv2.cvtColor(face_boundingbox_bgr, cv2.COLOR_BGR2RGB)
            #Приводим к нужному формату
            img = image.array_to_img(face_boundingbox_rgb)
            img = img.resize([224,224])
            img = np.array([np.array(img)])
            #Предсказание модели
            res_class = self.predict(img)
            #Добавляем bounding_box + номер класса (потом будет название класса)
            rgb_image_with_boundingbox = cv2.rectangle(rgb_image_with_boundingbox, (x, y), (x + h, y + w), (0, 255, 0), 3)
            rgb_image_with_boundingbox = cv2.putText(rgb_image_with_boundingbox, res_class, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        return rgb_image_with_boundingbox
    def return_video(self):
        # здесь мы в бесконечном цикле считываем кадр с камеры и выводим его, используя cv2.imshow()
        # корректная остановка окна с камерой произойдет, когда мы нажмем q на клавиатуре 
        while(True):
            cv2.imshow("facial emotion recognition", self.add_bounding_box())  
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
    
```

```python
#Создаем объект класса
obj  = emotion_on_camera()
```

```python
#Вывод изображения с эмоцией
plt.imshow(obj.add_bounding_box())
```

```python
#Вывод видео 
obj.return_video()
```
